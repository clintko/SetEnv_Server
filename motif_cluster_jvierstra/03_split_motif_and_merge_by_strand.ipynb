## List motif clusters for each chromosome

**TEST**

**RUN**

```bash
FD_LOG=./log
mkdir -p ${FD_LOG}

sbatch -p scavenger \
    --array 0-23 \
    --mem 8G \
    -o ${FD_LOG}/hg38_archetype_list_motif_clusters.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh
CHROMS=($(seq 1 22) X Y)
CHROM=${CHROMS[${SLURM_ARRAY_TASK_ID}]}

### set input
FD_BED=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1
FN_BED=chr${CHROM}_rm_mouse.bed.gz

### set output
FD_OUT=${FD_BED}/chr${CHROM}
FN_OUT=clusters.txt

### print start message
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Input  file:       " ${FD_BED}/${FN_BED}
echo "Output file:       " ${FD_BED}/${FN_OUT}
echo
echo "Show the first few lines of the input file"
echo ${FD_BED}/${FN_BED}
zcat ${FD_BED}/${FN_BED} | head -3
echo

### filter motifs that have a MOUSE label
echo "RUN: Get all motif clusters in each chromosome"
mkdir -p ${FD_BED}/chr${CHROM}
zcat ${FD_BED}/${FN_BED} | cut -f4 | sort | uniq > ${FD_OUT}/${FN_OUT}
echo "Done!"
echo

### print end message
echo "Show the first few lines of the output file"
echo ${FD_OUT}/${FN_OUT}
cat -n ${FD_OUT}/${FN_OUT}
EOF
```

**RUN**

```bash
source ../load_module_bedtools.sh
FD_LOG=./log

sbatch -p scavenger \
    --array 0-23 \
    --mem 8G \
    -o ${FD_LOG}/hg38_archetype_split_motif_clusters.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh
CHROMS=($(seq 1 22) X Y)
CHROM=${CHROMS[${SLURM_ARRAY_TASK_ID}]}

### set input and output
FD_BED=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1
FN_BED=chr${CHROM}_rm_mouse.bed.gz
FD_OUT=${FD_BED}/chr${CHROM}

### print start message
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Input  file:       " ${FD_BED}/${FN_BED}
echo
echo "Show the first few lines of the input file"
echo ${FD_BED}/${FN_BED}
zcat ${FD_BED}/${FN_BED} | head -3
echo

### loop through each motif cluster and output the corresponding bed file
echo "RUN: loop through each motif cluster; filter and output to a bed file"
N_LINE=$(wc -l < ${FD_OUT}/clusters.txt)
for idx in $(seq 1 ${N_LINE}); do
    ### get motif and create output filename
    MOTIF=$(awk "NR==${idx}" ${FD_OUT}/clusters.txt)
    FNAME=$(echo ${MOTIF} | sed -e 's/\//_/g')
    printf "%+10s  %+10s\n" ${MOTIF} ${FNAME}
    
    ### filter to get each cluster
    zcat ${FD_BED}/${FN_BED} | awk -F $'\t' -v MOTIF=${MOTIF} '($4 == MOTIF)' > ${FD_OUT}/${FNAME}.bed
    
    ### merge the cluster annotation by strand
    bedtools merge -i ${FD_OUT}/${FNAME}.bed -c 4,5 -o distinct,mean > ${FD_OUT}/${FNAME}_merge.bed
done
echo "Done!"

EOF
```

## Merge motif clusters for each chromosome

**TEST**

```bash
CHROM=chr17
FD_MTF=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1
FD_BED=${FD_MTF}/${CHROM}
FN_TMP=tmp.bed
FN_OUT=${CHROM}_rm_mouse_merge.bed.gz

for FP_BED in $(ls ${FD_BED}/*merge*); do
    echo ${FP_BED}
done
```

**RUN**

```bash
source ../config_duke.sh
source ../load_module_bedtools.sh
FD_LOG=./log

sbatch -p scavenger \
    --array 0-23 \
    --mem 20G \
    -o ${FD_LOG}/hg38_archetype_merge_motif_clusters.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh
CHROMS=($(seq 1 22) X Y)
CHROM=chr${CHROMS[${SLURM_ARRAY_TASK_ID}]}

### set input and output
FD_MTF=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1
FD_BED=${FD_MTF}/${CHROM}
FN_TMP=tmp.bed
FN_OUT=${CHROM}_rm_mouse_merge.bed.gz

### print start message
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Output file:       " ${FD_MTF}/${FN_OUT}
echo

### combine all motif clusters into a single file
echo "RUN: loop through each motif cluster; combine to a single bed file"
cat /dev/null > ${FD_BED}/${FN_BED}
for FP_BED in $(ls ${FD_BED}/*merge*); do
    echo "Input file: " ${FP_BED}
    cat ${FP_BED} | head -3
    echo
    cat ${FP_BED} >> ${FD_BED}/${FN_TMP}
done

### sort and compress
echo "RUN: sort and compress"
bedtools sort -i ${FD_BED}/${FN_TMP} | gzip -cf > ${FD_MTF}/${FN_OUT}

### remove the temporary file
rm ${FD_BED}/${FN_TMP}

echo "Done!"
EOF
```
