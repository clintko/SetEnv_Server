## Remove motifs with MOUSE labeling

**RUN**

```bash
FD_LOG=./log
mkdir -p ${FD_LOG}

sbatch -p scavenger \
    --array 0-23 \
    --mem 8G \
    -o ${FD_LOG}/hg38_archetype_rm_mouse_motifs.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh
CHROMS=($(seq 1 22) X Y)
CHROM=${CHROMS[${SLURM_ARRAY_TASK_ID}]}

### set input
FD_BED=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1
FN_BED=chr${CHROM}.bed.gz

### set output
FN_OUT=chr${CHROM}_rm_mouse.bed.gz

### print start message
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Input  file:       " ${FD_BED}/${FN_BED}
echo "Output file:       " ${FD_BED}/${FN_OUT}
echo
echo "Show the first few lines of the input file"
echo ${FD_BED}/${FN_BED}
zcat ${FD_BED}/${FN_BED} | head -3
echo

### filter motifs that have a MOUSE label
echo "RUN: Filter motifs that have a MOUSE label"
zcat ${FD_BED}/${FN_BED} | grep -v --ignore-case MOUSE | gzip -cf > ${FD_BED}/${FN_OUT}
echo "Done!"
echo

### print end message
echo "Show the first few lines of the output file"
echo ${FD_BED}/${FN_OUT}
zcat ${FD_BED}/${FN_OUT} | head -3
EOF
```
