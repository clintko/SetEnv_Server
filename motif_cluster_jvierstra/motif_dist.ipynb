## Count motifs

```bash
source ../config_duke.sh
source ../load_module_bedtools.sh

sbatch -p ${NODE} \
    --array 0-23 \
    --mem 8G \
    --job-name "motif count" \
    -o ${FD_LOG}/motif_count_chrom.%a.txt \
    <<'EOF'
#!/bin/bash
### set environment
source ../config_duke.sh -v
CHROMS=($(seq 1 22) X Y)
CHROM=chr${CHROMS[${SLURM_ARRAY_TASK_ID}]}

FD_BED="${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1/${CHROM}"
FD_OUT="${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1/dist_${CHROM}"

###
FN_OUT="count_motif.txt"

cat /dev/null > ${FD_OUT}/${FN_OUT}
for FP_BED in $(ls ${FD_BED}/*merge.bed); do
    NUM=$(cat ${FP_BED} | wc -l) 
    MTF=$(basename ${FP_BED})
    echo -e "${CHROM}\t${MTF}\t${NUM}">> ${FD_OUT}/${FN_OUT}
done

###
FN_OUT="count_motif_filter95.txt"

cat /dev/null > ${FD_OUT}/${FN_OUT}
for FP_BED in $(ls ${FD_BED}/*merge.bed); do
    NUM=$(cat ${FP_BED} | awk '$5>10.81' | wc -l) 
    MTF=$(basename ${FP_BED})
    echo -e "${CHROM}\t${MTF}\t${NUM}">> ${FD_OUT}/${FN_OUT}
done

EOF
```

## Motif pair

```bash
source ../config_duke.sh
source ../load_module_bedtools.sh

sbatch -p ${NODE} \
    --array 0-23 \
    --mem 8G \
    --job-name "motif distance" \
    -o ${FD_LOG}/motif_distance_chrom.%a.txt \
    <<'EOF'
#!/bin/bash
### set environment
source ../config_duke.sh -v
CHROMS=($(seq 1 22) X Y)
CHROM=chr${CHROMS[${SLURM_ARRAY_TASK_ID}]}

MTF1=AP1_1
MTF2=NR_20

###
FD_BED="${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1/${CHROM}"
FN_BED_A="${MTF1}_merge.bed"
FN_BED_B="${MTF2}_merge.bed"
FD_OUT="${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1/closest_pair/${CHROM}"

### print start message
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo ++++++++++++++++++++++++++++++++++++++++++++++++
echo "Input  file: " ${FD_BED}/${FN_BED_A}
echo "Input  file: " ${FD_BED}/${FN_BED_B}
echo "Output file: " ${FD_OUT}/${FN_OUT}
echo
echo "Show the first few lines of the input file"
echo    ${FD_BED}/${FN_BED_A}
head -3 ${FD_BED}/${FN_BED_A}
echo    ${FD_BED}/${FN_BED_B}
head -3 ${FD_BED}/${FN_BED_B}
echo

### RUN
timer_start=`date +%s`
mkdir -p ${FD_OUT}

FN_OUT="${MTF1}_${MTF2}.bed"
bedtools closest -a ${FD_BED}/${FN_BED_A} -b ${FD_BED}/${FN_BED_B} -d > ${FD_OUT}/${FN_OUT}

FN_OUT="${MTF2}_${MTF1}.bed"
bedtools closest -a ${FD_BED}/${FN_BED_B} -b ${FD_BED}/${FN_BED_A} -d > ${FD_OUT}/${FN_OUT}


### print end message
echo
echo ++++++++++++++++++++++++++++++++++++++++++++++++
echo "Show the first few lines of the output file"
echo    ${FD_OUT}/${FN_OUT}
head -3 ${FD_OUT}/${FN_OUT}
echo

timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"

EOF
```
