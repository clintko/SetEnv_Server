**Initialization**

````bash
ls ${FD_SRC}/genome/hg38.fa.gz
````

````bash
FD_GEN=${FD_ANN}/genome/hg38

mkdir -p ${FD_GEN}/fasta
gunzip -c ${FD_SRC}/genome/hg38.fa.gz > ${FD_GEN}/fasta/hg38.fa
````

### Split the fasta file by chromosome

````bash
cd ${FD_GEN}/fasta
csplit -s -z hg38.fa '/>/' '{*}'
````

### Rename the file

````bash
cd ${FD_GEN}/fasta
for i in xx* ; do \
    n=$(sed 's/>// ; s/ .*// ; 1q' "$i") ; \
    mv "$i" "$n.fa" ; \
done
````

````bash
ls -lh ${FD_GEN}/fasta
````

### Compress all fasta files
`gzip *`
- This will zip each file individually and DELETE the original.
- Use -k (--keep) option to keep the original files.
- This may not work if you have a huge number of files due to limits of the shell

```
gzip ${FD_GEN}/fasta/*.fa
```

Notice that bedtools does not work with gzipped fasta file, so I comment out the zipping code.

### Set genome size

````bash
echo ${FD_GEN}
````

````bash
cp ${FD_SRC}/genome/hg38.chrom.sizes ${FD_GEN}/
cat ${FD_GEN}/hg38.chrom.sizes
````

**Filter the chromosome chr1-22,X,Y**

````bash
cat hg38.chrom.sizes |\
    grep -P 'chr[0-9XY]{1,2}\t' |\
    sort -k1,1 -V -s > ${FD_GEN}/hg38.chrom.sizes.rm
    
cat ${FD_GEN}/hg38.chrom.sizes.rm
````

**Split the file by chromosome**

````bash
for i in {1..22} X Y; do
    cat ${FD_GEN}/hg38.chrom.sizes | grep -P "chr${i}\t" > ${FD_GEN}/hg38.chrom.sizes.chr${i}
done
````
